```python
from autoscraper import AutoScraper
```

```python
url="https://www.amazon.in/s?k=mi+phone"
want_list=['Redmi 9A (Nature Green, 2GB RAM, 32GB Storage) | 2GHz Octa-core Helio G25 Processor | 5000 mAh Battery','₹7,499','145,347']
```

```python
scrap=AutoScraper()
data=scrap.build(url,want_list)
result=scrap.get_result_similar(url,grouped=True)
```

```python
result
```

```python
scrap.set_rule_aliases({'rule_u176': 'Product Title','rule_k5k3':'Price','rule_5a0q': 'No. of reviews'})
# keep same rule for other queries alos
scrap.keep_rules(['rule_u176','rule_k5k3','rule_5a0q'])
scrap.save('web-Scrape')
```

```python
result=scrap.get_result_similar('https://www.amazon.in/s?k=realme+phone',group_by_alias=True)
```
